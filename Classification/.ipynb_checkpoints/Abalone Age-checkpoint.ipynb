```python
import graphlab
import numpy as np
```

```python
abalone = graphlab.SFrame.read_csv('abalone.csv', header = False)
```

```python
abalone
```

```python
# let's rename the columns as in abalone.names.txt
abalone.rename({'X1':'Sex', 'X2':'Length', 'X3':'Diameter', 'X4':'Height', 'X5':'Whole weight',
                     'X6':'Shucked weight', 'X7':'Viscera weight', 'X8':'Shell weight', 'X9':'Rings'})
```

```python
# Rings is the age to be predicted
abalone['Rings'].unique().sort()
```

```python
# Create dummy variables for Sex
abalone['Male'] = abalone['Sex'] == 'M'
abalone['Female'] = abalone['Sex'] == 'F'
abalone['Infant'] = abalone['Sex'] == 'I'
abalone = abalone.remove_column('Sex')
abalone
```

```python
def get_numpy_data(data_sframe, features, label):
    data_sframe['intercept'] = 1 # add intercept feature
    features = ['intercept'] + features
    feature_sframe = data_sframe[features]
    feature_matrix = feature_sframe.to_numpy() # convert to NumPy matrix
    label_sarray = data_sframe[label]
    label_array = label_sarray.to_numpy() # convert to NumPy array
    return(feature_matrix, label_array)
```

```python
features = abalone.column_names()
features.remove('Rings')
```

```python
feature_matrix, rings = get_numpy_data(abalone, features, 'Rings')
```
